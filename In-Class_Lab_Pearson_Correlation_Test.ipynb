# **In-Class Lab: Pearson Correlation Test Using Real-World Advertising Data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import math
```

## Part 1: Data Familiarization

```python
df = pd.read_csv("Advertising.csv")
df.head()
```

## Identify research questions:
* Is there a significant relationship between **TV advertising** and **sales**?
* How does **radio advertising** correlate with **sales**?
* Does **newspaper advertising** have a strong correlation with **sales**?

## Part 2: Pearson Correlation Analysis

## Manual Calculation (Optional)

```python
# TV advertising and Sales
x1 = df["TV"].sum()
y1 = df["Sales"].sum()
x_y1 = (df["TV"] * df["Sales"]).sum()
sqrt_x1 = (df["TV"] ** 2).sum()
sqrt_y1 = (df["Sales"] ** 2).sum()

r_tv = ((len(df) * x_y1) - (x1 * y1)) / (math.sqrt(((len(df) * sqrt_x1) - (x1 ** 2)) * ((len(df) * sqrt_y1) - (y1 ** 2))))
```

```python
# Radio advertising and Sales
x2 = df["Radio"].sum()
y2 = df["Sales"].sum()
x_y2 = (df["Radio"] * df["Sales"]).sum()
sqrt_x2 = (df["Radio"] ** 2).sum()
sqrt_y2 = (df["Sales"] ** 2).sum()

r_radio = ((len(df) * x_y2) - (x2 * y2)) / (math.sqrt(((len(df) * sqrt_x2) - (x2 ** 2)) * ((len(df) * sqrt_y2) - (y2 ** 2))))
```

```python
# Newspaper advertising and Sales
x3 = df["Newspaper"].sum()
y3 = df["Sales"].sum()
x_y3 = (df["Newspaper"] * df["Sales"]).sum()
sqrt_x3 = (df["Newspaper"] ** 2).sum()
sqrt_y3 = (df["Sales"] ** 2).sum()

r_newspaper = ((len(df) * x_y3) - (x3 * y3)) / (math.sqrt(((len(df) * sqrt_x3) - (x3 ** 2)) * ((len(df) * sqrt_y3) - (y3 ** 2))))
```

```python
print(f"Correlation (TV vs Sales): {r_tv}")
print(f"Correlation (Radio vs Sales): {r_radio}")
print(f"Correlation (Newspaper vs Sales): {r_newspaper}")
```

## Python-Based Calculation

```python
# Pearson correlation between TV advertising and Sales
corr_tv_sales, _ = stats.pearsonr(df["TV"], df["Sales"]) 

# Pearson correlation between Radio advertising and Sales
corr_radio_sales, _ = stats.pearsonr(df["Radio"], df["Sales"]) 

# Pearson correlation between Newspaper advertising and Sales
corr_news_sales, _ = stats.pearsonr(df["Newspaper"], df["Sales"]) 

# Print the result
print(f"Pearson Correlation (TV vs Sales): {corr_tv_sales}")
print(f"Pearson Correlation (Radio vs Sales): {corr_radio_sales}")
print(f"Pearson Correlation (Newspaper vs Sales): {corr_news_sales}")
```

### Interpretation
* **Pearson Correlation Coefficient (r=0.7822)**: This indicates a **strong positive** linear relationship between **TV advertising and Sales**.
* **Pearson Correlation Coefficient (r=0.5762)**: This indicates a **moderate positive** linear relationship between **Radio advertising and Sales**.
* **Pearson Correlation Coefficient (r=0.2282)**: This indicates a **weak positive** linear relationship between **Newspaper advertising and Sales**.
