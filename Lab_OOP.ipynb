## Problem 1: Basic Class Definition

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        
    def car_info(self):
        return f"Year: [{self.year}], Make: [{self.make}], Model: [{self.model}]"
```

```python
info = Car("Toyota", "Corolla", 2020)
print(info.car_info())
```

## Problem 2: Class with Methods

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
        
    def get_area(self):
        return self.length * self.width
        
    def get_perimeter(self):
        return 2 * (self.length + self.width)
```

```python
rec = Rectangle(10, 5)
print(rec.get_area())
print(rec.get_perimeter())
```

## Problem 3: Class with Encapsulation

```python
class Account:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.__balance = balance
    
    def deposit(self, add_amount):
        self.__balance = self.__balance + add_amount
        return self.__balance
        
    def withdraw(self, withdraw_amount):
        if self.__balance - withdraw_amount < 0:
            print("There is not enough money to withdraw.")
        else:
            self.__balance = self.__balance - withdraw_amount
            return self.__balance
    
    def get_balance(self):
        return self.__balance
```

```python
bank_info = Account(98765432, 2000)
print(bank_info.deposit(300))
print(bank_info.withdraw(2000))
print(bank_info.get_balance())
```

## Problem 5: Class with List Attribute

```python
class Team:
    def __init__(self):
        self.__members = []
    
    def add_member(self, new):
        return self.__members.extend(new)
            
    def remove_member(self, remove):
        if remove in self.__members:
            self.__members.remove(remove)
    
    def get_members(self):
        return self.__members
```

```python
team_list = Team()
# add new members
team_list.add_member(["Alice", "Bob", "Charlie"])
print(team_list.get_members())

# remove a member
team_list.remove_member("Bob")
print(team_list.get_members())
```

## Problem 6: Class with Multiple Methods

```python
class Bank:
    def __init__(self, name):
        self.name = name
        self.accounts = []
        
    def add_account(self, add_acc):
        return self.accounts.extend(add_acc)
        
    def remove_accounts(self, rmv_acc):
        return self.accounts.remove(rmv_acc)
    
    def get_accounts(self):
        return self.accounts
    
    def bank_info(self):
        return f"Bank: [{self.name}], Account: {self.accounts}"
```

```python
info = Bank("Bank of Python")
# add bank accounts
info.add_account(["1111", "2222", "3333"])
print(info.get_accounts())

# remove bank account
info.remove_accounts("2222")
print(info.get_accounts())

# print the bank infomation
print(info.bank_info())
```

## Problem 7: Class with Boolean Methods

```python
class User:
    def __init__(self, username, password):
        self.username = username
        self.password = password
        
    def check_password(self, password):
        if password == self.password:
            return True
        else:
            return False
```

```python
user_info = User("user1", "pass123")
user_info.check_password("pass123")
```

## Problem 8: Class with Initialization Logic
