# **In-Class Lab: Mean-Based Z-Test Questions**

```python
import numpy as np
import scipy.stats as stats
```

## Questiion 1: Average Time on a Website
### Null Hypothesis: Average time on Website = 10
### Alternative Hypothesis: Average time on Website != 10

```python
population_mean = 10
sample_mean = 9.5
sample_std = 1.8
sample_size = 50
significance_level = 0.05

z = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
p_value = (1 - stats.norm.cdf(abs(z))) * 2
print(f"z_score: {z}")
print(f"p_value: {p_value}")

if p_value < significance_level:
    print("Reject the null hypothesis: The sample average time on website is significantly different from the claimed average.")
else:
    print("Fail to reject the null hypothesis: The sample average time on website is not significantly different from the claimed average.")
```
